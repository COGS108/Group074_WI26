```python
# Define feature categories
counting_stats = ['pts', 'fg', 'fga', 'x3p', 'x3pa', 'x2p', 'x2pa',
                  'ft', 'fta', 'orb', 'drb', 'trb', 'ast', 'stl',
                  'blk', 'tov', 'pf', 'mp', 'g', 'gs', 'trp_dbl']

efficiency_stats = ['fg_percent', 'x3p_percent', 'x2p_percent',
                    'ft_percent', 'e_fg_percent']

# Combine all features for PCA
pca_features = counting_stats + efficiency_stats

print(f"Total features for PCA: {len(pca_features)}")
print(f"\nCounting stats ({len(counting_stats)}): {counting_stats}")
print(f"\nEfficiency stats ({len(efficiency_stats)}): {efficiency_stats}")

# Extract feature matrix
X = final_df[pca_features].copy()
print(f"\nFeature matrix shape: {X.shape}")
```

```python
# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Verify standardization
print("Verification of standardization:")
print(f"Mean of each feature (should be ~0): {X_scaled.mean(axis=0).round(10)[:5]}...")
print(f"Std of each feature (should be ~1): {X_scaled.std(axis=0).round(2)[:5]}...")
```

```python
# Fit PCA with all components
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Calculate variance explained
variance_explained = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(variance_explained)

# Display results
print("PCA Variance Explained by Component:")
print("-" * 50)
for i, (var, cum_var) in enumerate(zip(variance_explained[:10], cumulative_variance[:10])):
    print(f"PC{i+1}: {var:.4f} ({var*100:.2f}%) | Cumulative: {cum_var*100:.2f}%")

# Determine components needed for thresholds
n_80 = np.argmax(cumulative_variance >= 0.80) + 1
n_85 = np.argmax(cumulative_variance >= 0.85) + 1
n_90 = np.argmax(cumulative_variance >= 0.90) + 1

print(f"\nComponents needed for 80% variance: {n_80}")
print(f"Components needed for 85% variance: {n_85}")
print(f"Components needed for 90% variance: {n_90}")
```
